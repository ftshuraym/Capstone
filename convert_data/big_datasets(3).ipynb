```python
"""
  Loading the data from additional files data folder.
"""

import pandas as pd

"""
  Reading the txt files.
"""

mainDir = ''
unique_tracks = pd.read_csv(mainDir + "MillionSongSubset/AdditionalFiles/subset_unique_tracks.txt",sep="<SEP>",names=['track_id','song_id','artist_name','song_title'])
artist_ids = pd.read_csv(mainDir + "MillionSongSubset/AdditionalFiles/subset_unique_artists.txt",sep="<SEP>",names=['artist_id','artist_mbid','track_id','artist_name'])
artist_terms = pd.read_csv(mainDir+"MillionSongSubset/AdditionalFiles/subset_unique_terms.txt",names=['terms'])
artist_musicbrainz_tags = pd.read_csv(mainDir+"MillionSongSubset/AdditionalFiles/subset_unique_mbtags.txt",names=['tags'])
total_tracks = pd.read_csv(mainDir + "MillionSongSubset/AdditionalFiles/subset_tracks_per_year.txt",sep="<SEP>",names=['year','track_id','artist_name','song_title'])
artist_locations = pd.read_csv(mainDir + "MillionSongSubset/AdditionalFiles/subset_artist_location.txt",sep="<SEP>",names=['artist_id','lattiude','longitude','track','location'])

"""
  Reading .db files
"""

import sqlite3
import pandas as pd

def get_data(filepath, table_name):
    cnx = sqlite3.connect(filepath)
    df = pd.read_sql_query("SELECT * FROM " + table_name, cnx)
    return df

tracks_metadata = get_data(mainDir + 'MillionSongSubset/AdditionalFiles/subset_track_metadata.db', 'songs')
artist_id_tracks = get_data(mainDir+'MillionSongSubset/AdditionalFiles/subset_artist_term.db', 'artist_term')
artist_similarity = get_data(mainDir+'MillionSongSubset/AdditionalFiles/subset_artist_similarity.db', 'similarity')
```

```python
"""
  Loading data from h5 type files.
"""
import pandas as pd
import tables
import os
#import progressbar

"""
  Defining functions to get all required data from h5 file.
"""

def get_artist_mbid(h5):
      return h5.root.metadata.songs[0][8]

def get_artist_mbtags(h5):
      return h5.root.musicbrainz.artist_mbtags.shape[0]

def get_artist_mbtags_count(h5):
      return h5.root.musicbrainz.artist_mbtags_count.shape[0]

def get_artist_name(h5):
      return h5.root.metadata.songs[0][9]

def get_artist_playmeid(h5):
      return h5.root.metadata.songs[0][10]

def get_artist_terms(h5):
      return h5.root.metadata.artist_terms.shape[0]

def get_artist_terms_freq(h5):
      return h5.root.metadata.artist_terms_freq.shape[0]

def get_artist_terms_weight(h5):
      return h5.root.metadata.artist_terms_weight.shape[0]

def get_audio_md5(h5):
      return h5.root.analysis.songs[0][1]

def get_key_confidence(h5):
      return h5.root.analysis.songs[0][22]

def get_loudness(h5):
      return h5.root.analysis.songs[0][23]

def get_release(h5):
      return h5.root.metadata.songs[0][14]

def get_release_7digitalid(h5):
      return h5.root.metadata.songs[0][15]

def get_title(h5):
      return h5.root.metadata.songs[0][18]

def get_track_7digitalid(h5):
      return h5.root.metadata.songs[0][19]

def get_track_id(h5):
      return h5.root.analysis.songs[0][30]

def get_year(h5):
      return h5.root.musicbrainz.songs[0][1]

# get column data for a row
def transform_to_row(h5):
    row_data = [
                  get_artist_mbid(h5),
                  get_artist_mbtags(h5),
                  get_artist_mbtags_count(h5),
                  get_artist_name(h5),
                  get_artist_playmeid(h5),
                  get_artist_terms(h5),
                  get_artist_terms_freq(h5),
                  get_artist_terms_weight(h5),
                  get_audio_md5(h5),
                  get_key_confidence(h5),
                  get_loudness(h5),
                  get_release(h5),
                  get_release_7digitalid(h5),
                  get_title(h5),
                  get_track_7digitalid(h5),
                  get_track_id(h5),
                  get_year(h5)]
    return row_data

  # getting all files list of h5
def get_all_files_list(baseDir):
    all_files = []
    for root, dirs, files in os.walk(baseDir):
        for file in os.listdir(root):
            if file.endswith('.h5'):
                all_files.append(os.path.join(root,file))
    return all_files
```

```python
# Creating empty dataframe and getting row data.

baseDir = 'MillionSongSubset/data/'
total_files_list = get_all_files_list(baseDir)

columns_dataframe = [
              'artist_mbid',
              'artist_mbtags',
              'artist_mbtags_count',
              'artist_name',
              'artist_playmeid',
              'artist_terms',
              'artist_terms_freq',
              'artist_terms_weight',
              'audio_md5',
              'key_confidence',
              'loudness',
              'release',
              'release_7digitalid',
              'title',
              'track_7digitalid',
              'track_id',
              'year'
]

total_df = pd.DataFrame(columns=columns_dataframe)
```

```python
# Filling the dataframe with every song's data

for index, file in enumerate(total_files_list):
    h5 = tables.open_file(file, mode='r')
    total_df.loc[index] = transform_to_row(h5)
    h5.close()
    print("Conpleted " + str(index) + " out of "+ str(len(total_files_list)))
```
